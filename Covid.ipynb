```python
import pandas as pd
%matplotlib inline
f_cases  = "time_series_covid19_confirmed_global_narrow.csv"
f_deaths = "time_series_covid19_deaths_global_narrow.csv"
#f_recov  = "time_series-ncov-Recovered.csv"
```

### Analysis of data about confirmed cases

```python
# Data read from file
df_cases = pd.read_csv(f_cases)
pd.options.display.max_rows = 125
df_cases = df_cases[1:]    # skip 1st line
#df_cases
```

```python
# conversion of Value to integer numbers
df_cases['Value'] = df_cases['Value'].fillna(0).astype(int)
# fillna is to replace NaN by 0
```

```python
df_cases
# To list all districts of China
display(df_cases.loc[df_cases['Country/Region']=='China']['Province/State'].drop_duplicates())

# To look at Hubei district of China
df_cases.loc[df_cases['Province/State']=='Hubei']
```

```python
# Build a Series with the list of countries
countrySeries = df_cases['Country/Region'].drop_duplicates().reset_index(drop=True)
pd.options.display.max_rows = 185
print(countrySeries)
```

### Plot construction
For the purpose of making of plot of the evolution of confirmed cases per country we need a dataframe with dates as index, and the columns titles are countries, rows carry the confirmed cases values.

##### STEP 1: Build the total dataframe holding all countries
The *State/Province* records must be merged and the **Value** summed up to a unique record per country.

```python
first = True    # need to build a first DataFrame on which to merge the following DF.
second = False  # need to keep record of the second country for final column renaming.
#print(countrySeries[:50])
for p in countrySeries:
    if first:
        df_ = df_cases.loc[df_cases['Country/Region']==p]
        df_.reset_index(inplace=True)
        df_ = pd.DataFrame(df_, columns=['Date','Value'])
        df_ = df_.groupby('Date').sum()
        df_cases_par_pays = df_
        first=False
        second=True
        first_country = p
    else:
        df_ = df_cases.loc[df_cases['Country/Region']==p]
        df_ = pd.DataFrame(df_, columns=['Date','Value'])
        df_ = df_.groupby('Date').sum()
    
        df_cases_par_pays = pd.merge(df_cases_par_pays,df_, on='Date', how='outer')
        df_cases_par_pays = df_cases_par_pays.rename(columns={'Value': p})
        if second :
            second_country = p
            second = False

df_cases_par_pays = df_cases_par_pays.rename(columns={'Value_x': first_country,'Value_y': second_country})
#display(df_cases_par_pays)
```

##### STEP 2:  Keep most impacted countries

```python
# Extraction of most impacted countries - choose date to consider and # of top countries
n_top_countries = 7
at_date = '2020-04-06'

tmp = df_cases_par_pays.loc[at_date].sort_values(ascending=False)
s_most_confirmed = tmp[:n_top_countries]
s_countries_most_confirmed = s_most_confirmed.keys()
print(s_countries_most_confirmed)
```

```python
df_most_confirmed = pd.DataFrame(df_cases_par_pays, columns=s_countries_most_confirmed)
#display(df_most_confirmed)
ax=df_most_confirmed.plot(figsize=(12,10), colormap='jet', lw=2, marker='.', markersize=10, \
                          legend=True)
ax.set_title(str(n_top_countries)+" most contaminated countries",fontsize= 20)
ax.set_xlabel("Date", fontsize=16)
ax.tick_params(axis='x', labelrotation=30 )
ax.set_ylabel("# of confirmed contaminated cases", fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=14)
ax.legend(fontsize=14)
```

### Analysis of deaths
 1. construction of plot with number of deaths versus dates. Plot for each country overlayed.
 2. construction of plot with "number of deaths / confirmed cases" versus dates. Plot for each country overlayed.

```python
df_deaths = pd.read_csv(f_deaths)
pd.options.display.max_rows = 25
df_deaths = df_deaths[1:]
#df_deaths
```

```python
# conversion of Value to integer numbers
df_deaths['Value'] = df_deaths['Value'].fillna(0).astype(int)
# fillna is to replace NaN by 0
```

```python
df_deaths
```

### 1. Plot construction
For the purpose of making of plot of the evolution of confirmed cases per country we need a dataframe with dates as index, and the columns titles are countries, rows hold the confirmed cases values.

##### STEP 1: Build the total dataframe holding all countries

```python
first  = True   # need to build a first DataFrame on which to merge the following DF.
second = False  # need to keep record of the second country for final column renaming.
#print(countrySeries[:50])
for p in countrySeries:
    if first:
        df_ = df_deaths.loc[df_deaths['Country/Region']==p]
        df_.reset_index(inplace=True)
        df_ = pd.DataFrame(df_, columns=['Date','Value'])
        df_ = df_.groupby('Date').sum()
        df_deaths_par_pays = df_
        first=False
        second=True
        first_country = p
    else:
        df_ = df_deaths.loc[df_deaths['Country/Region']==p]
        df_ = pd.DataFrame(df_, columns=['Date','Value'])
        df_ = df_.groupby('Date').sum()
    
        df_deaths_par_pays = pd.merge(df_deaths_par_pays,df_, on='Date', how='outer')
        df_deaths_par_pays = df_deaths_par_pays.rename(columns={'Value': p})
        if second :
            second_country = p
            second = False

df_deaths_par_pays = df_deaths_par_pays.rename(columns={'Value_x': first_country,'Value_y': second_country})
#display(df_cases_par_pays)
```

```python
# Extraction of most impacted countries - choose date to consider and # of top countries
n_top_countries = 7
at_date = '2020-04-06'

tmp = df_deaths_par_pays.loc[at_date].sort_values(ascending=False)
s_most_deaths = tmp[:n_top_countries]
s_countries_most_deaths = s_most_deaths.keys()
print(s_countries_most_deaths)
```

```python
df_most_deaths = pd.DataFrame(df_deaths_par_pays, columns=s_countries_most_deaths)
ax=df_most_deaths.plot(figsize=(12,10), colormap='rainbow', lw=2, marker='.', markersize=10, legend=True)
ax.set_title(str(n_top_countries)+" most impacted countries by patients death",fontsize= 20)
ax.set_xlabel("Date", fontsize=16)
ax.tick_params(axis='x', labelrotation=30 )
ax.set_ylabel("# of persons died", fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=14)
ax.legend(fontsize=10)
```

```python
df_cases_deaths_par_pays = pd.merge(df_cases_par_pays, df_deaths_par_pays, on='Date', how='outer')
df_cases_deaths_par_pays.reset_index(inplace=True)
#display(df_cases_deaths_par_pays)
#print(df_cases_deaths_par_pays['France_x'],"",df_cases_deaths_par_pays['France_y'])
df_ratio_death_case = pd.DataFrame(df_cases_deaths_par_pays['Date'])
for i in s_countries_most_confirmed:
    df_ratio_death_case[i] = (df_cases_deaths_par_pays[i+'_y']/df_cases_deaths_par_pays[i+'_x'])*100
df_ratio_death_case = df_ratio_death_case.fillna(0)
df_ratio_death_case = df_ratio_death_case.set_index('Date')
#display(df_ratio_death_case)    
```

```python
# Plot 
ax=df_ratio_death_case.plot(figsize=(12,10), colormap='jet', lw=2, marker='.', \
                            markersize=10, legend=True, ylim=[0.,30.])
ax.set_title(str(n_top_countries)+" most impacted countries : ratio of deaths / cases",fontsize= 20)
ax.set_xlabel("Date", fontsize=16)
ax.tick_params(axis='x', labelrotation=30 )
ax.set_ylabel("ratio (%)", fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=14)
ax.legend(fontsize=14)
```
